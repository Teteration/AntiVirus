## Data pipeline

```python
# Step 1: Get string input from user
user_input = input("Enter your question: ")

# Step 2: Split the input by space
words = user_input.split()

# Step 3: Remove question marks from the words
words = [word.replace('?', '') for word in words]

# Step 4: Create the URL for semanticscholar.org
search_query = '%20'.join(words)
query_url = f"https://www.semanticscholar.org/search?q={search_query}&sort=relevance&page=1"

# Print the output URL
print("Output:", query_url)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.semanticscholar.org/api/1/search"


headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/119.0',
    'Accept': '*/*',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate',
    'Content-Type': 'application/json',
    'X-S2-Ui-Version': '036cd44a3c45c5db1a858fb7c99be4a91035be51',
    'X-S2-Client': 'webapp-browser'
}



# Define the payload
payload = {
    "queryString": "",
    "page": 1,
    "pageSize": 10,
    "sort": "relevance",
    "authors": [],
    "coAuthors": [],
    "venues": [],
    "yearFilter": None,
    "requireViewablePdf": False,
    "fieldsOfStudy": [],
    "hydrateWithDdb": True,
    "includeTldrs": True,
    "performTitleMatch": True,
    "includeBadges": True,
    "getQuerySuggestions": False,
    "cues": [],
    "includePdfVisibility": False
}
```

```python
data = []
# Note: its maybe that there are less that 100 results or No results for a specific query
for i in range(1,11):

  payload["page"] = i
  payload["queryString"] = user_input


  response = requests.post(url, headers=headers, json=payload)
  res = response.json()
  # print("res",res)
  try:
    for j in range(0,10):
      data.append(res["results"][j])
  except:
    print("its maybe that there are lesser that 100 results or No results for this specific query")
# Process the response data as needed
print(data)
  # print(res)
```

```python
from sentence_transformers import SentenceTransformer

retriever = SentenceTransformer(
    "flax-sentence-embeddings/all_datasets_v3_mpnet-base"
)
retriever
```

```python
from tqdm.auto import tqdm

batch_size = 5

for i in tqdm(range(0, len(data), batch_size)):

  i_end = min(i+batch_size, len(df))

  batch = df.iloc[i:i_end]

  emb = retriever.encode(batch['paperAbstract'].tolist()).tolist()

  meta = batch.to_dict(orient="records")

  ids = [f"{idx}" for idx in range(i, i_end)]

  to_upsert = list(zip(ids, emb, meta))
  print(to_upsert)

  _ = index.upsert(vectors=to_upsert)


index.describe_index_stats()
```

## Q&A model

####abstractive

```python
from transformers import BartTokenizer, BartForConditionalGeneration


tokenizer = BartTokenizer.from_pretrained('vblagoje/bart_lfqa')
generator = BartForConditionalGeneration.from_pretrained('vblagoje/bart_lfqa')
```

```python
 def query_pinecone(query, top_k):

  xq = retriever.encode([query]).tolist()

  xc = index.query(xq, top_k=top_k, include_metadata=True)
  return xc
```

```python
def format_query(query, context):

  context = [f"<P> {m['metadata']['paperAbstract']}" for m in context]

  context = " ".join(context)

  query = f"question: {query} context: {context}"
  return query
```

یه کد بزنم که رنک همه داکیومنت هارو با مشخض کنه

```python
query = user_input
result = query_pinecone(query, top_k=2)
result
```

```python
from pprint import pprint

query = format_query(query, result['matches'])
query
```

```python
def generate_answer(query):

  inputs = tokenizer([query], max_length=1024, return_tensors="pt")

  ids = generator.generate(
      inputs["input_ids"], num_beams=2, min_length=20, max_length=40
  )
  answer = tokenizer.batch_decode(
      ids, skip_special_tokens=True, clean_up_tokenization_spaces=False
  )[0]

  return pprint(answer)
```

```python
generate_answer(query)
```
